## Create pie chart:-

```python
df_pie = [df_merged["VeryActiveMinutes"], df_merged["FairlyActiveMinutes"], df_merged["LightlyActiveMinutes"]]
```

```python
pie_df = pd.DataFrame(data = df_pie)
```

```python
pie_df["Results"] = pie_df.sum(axis=1)
```

```python
pie_df
```

```python
pie_dflist = pie_df["Results"]
pie_dflist = pd.DataFrame(data=pie_dflist)
```

```python
pie_dflist
```

```python
pie_dflist.plot(kind = "pie",y='Results', title="Daily Activity Lavels", legend=False, \
                   autopct='%1.1f%%', explode=(0, 0, 0.1), \
                   shadow=True, startangle=0)
```

* This above pie chart is clearly showing that 83% participants are lightly active which is more then majority.
So, At least 150 minutes a week of moderate intensity activity such as brisk walking. At least 2 days a week of activities that strengthen muscles. Activities to improve balance such as standing on one foot. Aim for the recommended activity level but be as active as one is able.
